## Age and Duration of ride in minutes

> Age of customers and duration of ride in minutes can be used to better understand how the FordGo bike system operates and works. If improvements are needed to improve number of rides for different age groups. From the plot, it shows that customers within the age brackets of 20 to 85 usually have duration of rides that last less than 100 minutes. With the duration of rides reducing drastically with age groups less than 60. The FordGo bike system have a handful of users that are above 80, which means that if they are to see an increase in this number they are to provide special services to the age groups.


```python
# plotting relation between ages and duration time 
plt.scatter(data=df, x='duration_min' ,y='Age', alpha=5/20);
plt.xlabel('Duration Time (minutes)')
plt.ylabel('Ages');
plt.title('Distribution of Age and Time of Trip');
```

## Start hour of Rides and Gender
> The start hour represents the hour a user takes a ride, which was gotten from the start time variable in our dataset. The gender represents the gender type of users. The plot shows that across the different genders, the highest start hours of rides are usually 8am, 9am, 4pm, 5pm and 6pm. This could mean that the users of the system are usually workers or students that go out by 8am - 9am and close by 5pm - 6pm. And in the distribution it was observed that male genders had a higher count. FordGo bikek system can carry out a survey in order to improve the number of female users, the reasons might be comfort. Also, FordGo bike system can make more bikes available during these high hours in order to meet demands.


```python
sb.countplot(data = df, x = 'hour_start', hue = 'member_gender', palette = 'Greens');
plt.legend(ncol = 2);
plt.title('Gender and Start Hour of Rides');
plt.xlabel('Start hour');
```

## Distribution of Routes Longest Hours Duration

> Routes is the destination(end station name) from the starting(start station name) points, this plots depicts the total average time of the journey in hours. The longest route average time is from 11th St at Natoma St to 16th St Depot station which takes an average of 23 hours, followed Scott St at Golden Gate Ave to McAllister St at Baker St station with an average of 20 hours time duration. FordGo bike system can increase the number of bikes that ply these routes in order to generate more revenue.

```python
dd = pd.DataFrame(df.groupby(['start_station_name','end_station_name'])['duration_hr'].mean())
dd.sort_values(by='duration_hr', ascending=False).head(5).plot(kind='bar');
plt.title('Stations Route with Highest Average Duration Hour');
plt.ylabel('Average Duration (hours)');
plt.xlabel('Station Route');
```
